```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Embedding, Dense,LSTM
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
dataset = pd.read_csv('/content/train.csv')
```

```python
dataset
```

```python
df = dataset.dropna()
```

```python
X = df.drop('label',axis=1)
```

```python
y= df['label']
```

```python
X.shape
```

```python
y.shape
```

```python
voc_size=5000
```

```python
messages= X.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
LS= WordNetLemmatizer()
corpus=[]
```

```python
nltk.download('wordnet')
```

```python
PS = PorterStemmer()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_rep = [one_hot(word,voc_size) for word in corpus]
```

```python
onehot_rep
```

```python
length = 20
sequence_doc = pad_sequences(onehot_rep,padding='pre',maxlen=length)
```

```python
sequence_doc
```

```python
embedding_feature =20
model = Sequential()
model.add(Embedding(voc_size,embedding_feature,input_length=length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(sequence_doc)
```

```python
y.shape
```

```python
import numpy as np
X_final = np.array(sequence_doc)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X_final,y_final,test_size=0.3,random_state=43)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
# Assuming model is your trained Keras model
raw_predictions = model.predict(X_test)

# For binary classification, use a threshold (e.g., 0.5)
threshold = 0.5
binary_predictions = (raw_predictions > threshold).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,binary_predictions)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,binary_predictions)
```
